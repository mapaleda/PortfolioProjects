Every year, American high school students take SATs, which are standardized tests intended to measure literacy, numeracy, and writing skills. There are three sections - reading, math, and writing, each with a **maximum score of 800 points**. These tests are extremely important for students and colleges, as they play a pivotal role in the admissions process.

Analyzing the performance of schools is important for a variety of stakeholders, including policy and education professionals, researchers, government, and even parents considering which school their children should attend. 

You have been provided with a dataset called `schools.csv`, which is previewed below.

You have been tasked with answering three key questions about New York City (NYC) public school SAT performance.

```python
# Re-run this cell 
import pandas as pd # type: ignore

# Read in the data
schools = pd.read_csv("schools.csv")

# Preview the data
schools.head()
```

```python
# 1. Finding schools with the best math results

#Creating DataFrame and subsetting columns that will be included.
best_math_schools = pd.DataFrame(schools[["school_name", "average_math"]])

#Filtering schools that fulfill condition and sorting by said condition 
best_math_schools = best_math_schools[best_math_schools["average_math"] >= 800 * 0.8].sort_values("average_math", ascending=False)
best_math_schools
#schools.head()
```

```python
#2. Identifying the top 10 performing schools

#Creating column for average scores
schools["total_SAT"] = schools["average_math"] + schools["average_reading"] + schools["average_writing"]

#Creating Dataframe and subsetting the columns that will be included.
top_10_schools = pd.DataFrame(schools[["school_name", "total_SAT"]])

#Sorting by total scores and limiting to top 10 schools
top_10_schools = top_10_schools.sort_values("total_SAT", ascending=False).head(10)
top_10_schools
```

```python
# 3. Locating NYC borough with largest std dev in SAT performance

# Grouping data by borough
boroughs = schools.groupby("borough")["total_SAT"].agg(["count", "mean", "std"]).round(2)
boroughs

#Creating Dataframe
largest_std_dev = pd.DataFrame(boroughs)

#Filtering for the largest standard deviation
largest_std_dev = largest_std_dev[largest_std_dev["std"] == largest_std_dev["std"].max()]
largest_std_dev

#Renaming columns
largest_std_dev = largest_std_dev.rename(columns = {"count": "num_schools", "mean": "average_SAT", "std": "std_SAT"})
largest_std_dev
```
